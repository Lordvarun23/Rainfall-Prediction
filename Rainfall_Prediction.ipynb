# Data cleaning

```python
#Importing Libraries
```

```python
import numpy as np
```

```python
import pandas as pd
```

```python
#Reading the dataset
```

```python
data = pd.read_csv("austin_weather.csv")
```

```python
#Removing unnecessary coloumns
```

```python
data = data.drop(['Events', 'Date', 'SeaLevelPressureHighInches', 'SeaLevelPressureLowInches'], axis = 1)
```

```python
#Some Records in Pecipitaion has value 'T' which needs to be replaced with zero for better prediction
```

```python
data = data.replace('T', 0.0)
```

```python
#Some Records in dataset has value '-'which indicates null or nill.So it needs to be replaced with zero
```

```python
data = data.replace('-', 0.0)
```

```python
#saving the changes to the csv file
```

```python
data.to_csv('austin_final.csv')
```

# Training our model with Linear Regression

```python
#Importing Libraries
```

```python
import numpy as np
```

```python
import pandas as pd
```

```python
import sklearn as sk
```

```python
from sklearn.linear_model import LinearRegression
```

```python
import matplotlib.pyplot as plt
```

```python
#Importing cleaned dataset
```

```python
data=pd.read_csv("austin_final.csv")
```

```python
#getting the x values that is except precipitation rest columns will serve as x labels
```

```python
X = data.drop(['PrecipitationSumInches'], axis = 1)
```

```python
#the Y will be the output which is PrecipitationSumInches
```

```python
Y = data['PrecipitationSumInches']
```

```python
#converting it into 2-D Array
```

```python
Y = Y.values.reshape(-1, 1)
```

```python
#initialize a linear regression classifier
```

```python
clf=LinearRegression()
```

```python
#Training the classifier with our input data
```

```python
clf.fit(X, Y)
```

```python
#testing the model with our input
```

```python
inp = np.array([[74], [60], [45], [67], [49], [43], [33], [45],[57], [29.68], [10], [7], [2], [0], [20], [4], [31]])
```

```python
#converting the input array into 2-D array
```

```python
inp = inp.reshape(1, -1)
```

```python
#prediction for our input
```

```python
print('The precipitation in inches for the input is:', round(clf.predict(inp)[0][0],4))
```
